```python
%matplotlib inline
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with pd.HDFStore('results.h5') as store:
    perplexity = store.get('perplexity')
    coherence = store.get('coherence')
```

```python
perplexity.info()
```

```python
perplexity.vocab_size.value_counts().sort_index().plot.bar();
```

```python
fig, ax = plt.subplots(figsize=(14,8))
sns.heatmap(perplexity.groupby(['vocab_size', 'num_topics']).perplexity.median().unstack(), annot=True, fmt='.1f');
```

```python
# fig, ax = plt.subplots(figsize=(14,8))
sns.heatmap(perplexity.groupby(['min_df', 'max_df']).perplexity.median().unstack(), annot=True, fmt='.1f');
```

```python
sns.catplot(x='num_topics', y='perplexity', data=perplexity, hue='binary', col='passes', row='vocab_size');
```
